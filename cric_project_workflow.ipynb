```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class BatsmanState(TypedDict):
    runs:int
    balls:int
    fours:int
    sixes:int
    sr :float
    boundry_percent:float
    summary:str
    
```

```python
def calculate_sr(state:BatsmanState)->BatsmanState:
    runs = state['runs']
    balls = state['balls']
    sr = (runs/balls)*100
    return {'sr': sr}
```

```python
def calculate_bpb(state:BatsmanState)->BatsmanState:
    runs = state['runs']
    balls = state['balls']
    fours = state['fours']
    sixes = state['sixes']
    bpb = balls/(fours + sixes)
    return {'bpb': bpb}
```

```python
def calculate_boundry_percent(state:BatsmanState)->BatsmanState:
    fours = state['fours']
    sixes = state['sixes']
    runs = state['runs']
    boundry_percent = ((fours*4) + (sixes*6))/runs * 100
    return {'boundry_percent': boundry_percent}
```

```python
def summary(state:BatsmanState)->BatsmanState:
    runs = state['runs']
    balls = state['balls']
    fours = state['fours']
    sixes = state['sixes']
    sr = state['sr']
    boundry_percent = state['boundry_percent']
    
    print(f"Runs: {runs}")
    print(f"Balls: {balls}")
    print(f"Fours: {fours}")
    print(f"Sixes: {sixes}")
    print(f"Strike Rate: {sr}")
    print(f"Boundary Percentage: {boundry_percent}")
    
    return {'summary':summary}
```

```python
graph = StateGraph(BatsmanState)

graph.add_node('calculate_sr',calculate_sr)
graph.add_node('calculate_brp',calculate_bpb)
graph.add_node('calculate_boundry_percent',calculate_boundry_percent)
graph.add_node('summary',summary)

graph.add_edge(START, 'calculate_sr')
graph.add_edge(START, 'calculate_brp')
graph.add_edge(START, 'calculate_boundry_percent')

graph.add_edge('calculate_sr', 'summary')
graph.add_edge('calculate_brp', 'summary')
graph.add_edge('calculate_boundry_percent', 'summary')
graph.add_edge("summary",END)

workflow = graph.compile()
```

```python
intitial_state = BatsmanState(runs=120, balls=80, fours=10, sixes=5)
final_state = workflow.invoke(intitial_state)
```
