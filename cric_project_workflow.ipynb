```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class BatsmanState(TypedDict):
    runs:int
    balls:int
    fours:int
    sixes:int
    sr :float
    boundry_percent:float
    
```

```python
def calculate_sr(state:BatsmanState)->BatsmanState:
    runs = state['runs']
    balls = state['balls']
    state['sr'] = (runs/balls)*100
    return state
```

```python
def calculate_bpb(state:BatsmanState)->BatsmanState:
    runs = state['runs']
    balls = state['balls']
    fours = state['fours']
    sixes = state['sixes']
    state['bpb'] = balls/(fours + sixes)
    return state
```

```python
def calculate_boundry_percent(state:BatsmanState)->BatsmanState:
    fours = state['fours']
    sixes = state['sixes']
    runs = state['runs']
    state['boundry_percent'] = ((fours*4) + (sixes*6))/runs * 100
    return state
```

```python
def summary(state:BatsmanState)->BatsmanState:
    runs = state['runs']
    balls = state['balls']
    fours = state['fours']
    sixes = state['sixes']
    sr = state['sr']
    boundry_percent = state['boundry_percent']
    
    print(f"Runs: {runs}")
    print(f"Balls: {balls}")
    print(f"Fours: {fours}")
    print(f"Sixes: {sixes}")
    print(f"Strike Rate: {sr}")
    print(f"Boundary Percentage: {boundry_percent}")
    
    return state
```

```python
graph = StateGraph(BatsmanState)

graph.add_node('calculate_sr',calculate_sr)
graph.add_node('calculate_brp',calculate_brp)
graph.add_node('calculate_boundry_percent',calculate_boundry_percent)
graph.add_node('summary',summary)
```
